```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Constants
GRAVITY = 9.81  # gravitational acceleration (m/s^2)
DT = 0.1  # time step (s)
```

```python
# Initial conditions
x = 0  # initial position (m)
y = 100  # initial height (m)
vx = 10  # initial velocity in the x-direction (m/s)
vy = 0  # initial velocity in the y-direction (m/s)
```

```python
# Lists to store position and velocity values
x_values = []
y_values = []

while y > 0:
    # Update position and velocity
    x = x + vx * DT
    y = y + vy * DT
    vy = vy - GRAVITY * DT

    # Append values to lists
    x_values.append(x)
    y_values.append(y)
```

```python
# dataframe with both lists as columns
df = pd.DataFrame({'x': x_values, 'y': y_values})

# difference between the y coordinates
df['delta_y'] = df['y'].diff()
df
```

```python
# print(x_values)
# print(y_values)



# Plot position values
plt.plot(df['x'], df['y'])
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Position of falling ball')
plt.show()
```
